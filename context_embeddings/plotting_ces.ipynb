```python
import providedHelpers
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
```

```python
all_embeddings = []
for i in range(1768):
    all_embeddings.append(np.load(f'v_context_embeddings/{i}.npy'))
```

```python
pca_output = providedHelpers.pca_transform(all_embeddings)
pca_output
```

```python
# color stuff

word_set = set()
dat = pd.read_csv('normalized_AVG/hmongData-v-ST_AVG.csv')
for w in dat['Word']:
    word_set.add(w.split('_')[0])

used_colors = set()
color_map = {}
for word in word_set:
    new_color = tuple(np.random.choice(range(256), size=3))
    while new_color in used_colors:
        new_color = tuple(np.random.choice(range(256), size=3))
    color_map[word] = new_color
    used_colors.add(new_color)

color_map
```

```python
x = [x for x,_ in pca_output]
y = [y for _,y in pca_output]
colors = []
for w in dat['Word']:
    read_color = color_map[w.split('_')[0]]
    new_color = []
    for q in read_color:
        new_color.append(q/256)
    colors.append(new_color)

ax = plt.scatter(x, y, c = colors)
```

```python
labels = [word for word in color_map]

colors = []
for w in labels:
    read_color = color_map[w.split('_')[0]]
    new_color = []
    for q in read_color:
        new_color.append(q/256)
    colors.append(new_color)

handles = [mpatches.Patch(color=color, label=label) for label, color in zip(labels, colors)]
fig, ax = plt.subplots()

# Turn off both axes
ax.set_axis_off()
plt.legend(handles, labels, ncol=2)
```
