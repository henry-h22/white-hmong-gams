# Word Types
This code was used to explore the word types attested in the data on 8/8/2026.
All of the tones\* have a zipf-like distribution in terms of word types and the counts thereof, and the top six most common words for all tones have a fair amount of data, and seem to be in a class above the other word types. As such, I propose collapsing any further exploration into word type down to just these six, at least while we refine methodology and further explore the data.

\* not d, that one has two word types.

```python
import pandas as pd
import matplotlib.pyplot as plt
tones = ['b', 's', 'j', 'v', 'm', 'g', 'd', '0']
plt.rcParams["figure.figsize"] = (100,25)
plt.rcParams["font.size"] = 32
```

```python
for tone in tones:
    print(f'\n\n~~ Running Tone {tone} ~~')
    file = f'./normalized_AVG/hmongData-{tone}-ST_AVG.csv'
    df = pd.read_csv(file)
    word_counter = {}

    for index, row in df.iterrows():
        word = row.loc['Word'].split('_')[0]
        try:
            word_counter[word] += 1
        except KeyError:
            word_counter[word] = 1

    word_counter = dict(sorted(word_counter.items(), key=lambda item: item[1], reverse=True))

    print(word_counter)
    print(f'There are {len(word_counter)} word types attested for tone {tone}')

    most_common = list(word_counter)[:6]

    n = 'six' if len(most_common) == 6 else len(most_common)
    print(f'The {n} most common words of tone {tone} are: {most_common}')

    print(f'--- Graph of word type against log occurences for all words with tone {tone} ---')
    plt.bar(list(word_counter.keys()), word_counter.values(), color='g', width = 0.8)
    # plt.yscale('log') #! We use log scale just to make the graphs more level, but try re-running with this line commented to see the zipf distribution!
    plt.show()
```

Now that we've done that exploration, we'll use the following cells to add/clean up our "word" column so that our models have something to work with that's more thoughtfully prepared. i.e. we'll create a NormalizedWord column of the data frame that contains either the word (stripped of phrase information) or some "other" token (I'll use X) If the word in the column is one of the top 6 words of the tone, then the NormalizedWord will just be that word. Otherwise, X.

```python
for tone in tones:
    filename = f'hmongData-{tone}-ST_AVG.csv'
    loc = f'./normalized_AVG/{filename}'
    df = pd.read_csv(loc)

    # Step 1: Count the words and find the top 6 (same code as above, just again)
    word_counter = {}
    for index, row in df.iterrows():
        word = row.loc['Word'].split('_')[0]
        try:
            word_counter[word] += 1
        except KeyError:
            word_counter[word] = 1

    word_counter = dict(sorted(word_counter.items(), key=lambda item: item[1], reverse=True))
    most_common = list(word_counter)[:6]

    # Step 2: Go through and generate the normalized word for each column

    normalizedWord = []
    for index, row in df.iterrows():
        word = row.loc['Word'].split('_')[0]
        if word in most_common:
            normalizedWord.append(word)
        else: normalizedWord.append('X')

    df.insert(5, 'NormalizedWord', normalizedWord)
    df.to_csv(f'normWord/{filename}', index = False)
```
